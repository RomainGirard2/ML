```
all_RDD = sc.wholeTextFiles("FileStore/tables/city_data")

x1 = all_RDD.flatMapValues(lambda t: t.split("\r\n"))
x2 = x1.values()
x3 = x2.map(lambda t : int(t.split(" ")[1]))
seqOp = (lambda local_result, list_element: (local_result[0] + list_element,
local_result[1] + 1) )
combOp = (lambda some_local_result, another_local_result:
(some_local_result[0] + another_local_result[0], some_local_result[1] +
another_local_result[1]) )
avg_fr = x3.aggregate( (0, 0), seqOp, combOp)
print(avg_fr[0]/avg_fr[1])                              #Average monthly income of the shop in France
```
